```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('all_data.csv')
print(df.head())
print(df.info())
# Display basic statistics
print(df.describe())
```

```python
# prompt: i have four columns: "Country", a categorical with the country names, "Year" which is the year the entry is talking about, "GDP" whic is the gdp of that country that year, and "Life expectancy at birth "years" which is the life expectancy at birth from that country in years that year.
# help me do data analysis

import matplotlib.pyplot as plt
# Rename the column 'Life expectancy at birth (years)' to a simpler name
df = df.rename(columns={"Life expectancy at birth (years)":"LEABY"})

# Get unique countries
print(f"Unique countries: {df['Country'].unique()}")

# Explore average metrics per country
print("\nAverage metrics per country:")
print(df.groupby('Country')[['GDP', 'LEABY']].mean())

# Analyze relationship between GDP and Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP', y='LEABY', hue='Country')
plt.title('GDP vs Life Expectancy per Country')
plt.xlabel('GDP (in trillions)')
plt.ylabel('Life Expectancy at Birth (years)')
plt.show()

# Analyze trend of GDP over time for each country
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP Trend Over Time per Country')
plt.xlabel('Year')
plt.ylabel('GDP (in trillions)')
plt.grid(True)
plt.show()

# Analyze trend of Life Expectancy over time for each country
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='LEABY', hue='Country')
plt.title('Life Expectancy Trend Over Time per Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.grid(True)
plt.show()

# Distribution of Life Expectancy
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='LEABY', bins=10, kde=True)
plt.title('Distribution of Life Expectancy at Birth')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.show()

# Distribution of GDP
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='GDP', bins=10, kde=True)
plt.title('Distribution of GDP')
plt.xlabel('GDP (in trillions)')
plt.ylabel('Frequency')
plt.show()
```
